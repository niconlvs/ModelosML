##### <font color='264CC7'> Gráficos </font>

```python
# Definir las columnas a usar
numeric_cols = [
    'TransactionAmount',
    'CustomerAge',
    'LoginAttempts',
    'AccountBalance',
    ]

cat_cols = [
    'TransactionType',
    'Channel'
    ]
```

```python
# Estilo bonito para los gráficos
sns.set(style="whitegrid")

# Crear un histograma para cada columna numérica
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

```python
# Crear subplots
plt.figure(figsize=(15, 5))

for i, col in enumerate(cat_cols, 1):
    plt.subplot(1, 3, i)
    sns.countplot(data=df, x=col, palette='pastel', edgecolor='black')
    plt.title(f"Frecuencia de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.xticks(rotation=45)

plt.suptitle("Distribución de Variables Categóricas Seleccionadas", fontsize=16)
plt.tight_layout()
plt.show()
```

#### <font color='264CC7'> Escalar Datos </font>

```python
#eL DecisionTreeClassifier de scikit-learn no puede manejar texto sin procesar. Necesita que todas las variables independientes (X) sean numéricas.
df_encoded = pd.get_dummies(df, columns=['TransactionType', 'Channel'], drop_first=True)
```


### <font color='264CC7'> Entrenamiento del modelo y evaluación </font>

#### <font color='264CC7'> Entrenamiento del modelo </font>


```python
# Define X y y
x = df_encoded[['TransactionAmount', 'CustomerAge', 'LoginAttempts', 'AccountBalance'] +
               [col for col in df_encoded.columns if col.startswith('TransactionType_') or col.startswith('Channel_')]]
y = df['CustomerOccupation']

# Now split your data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize and train the Decision Tree model
model = DecisionTreeClassifier(random_state=76, criterion="entropy", max_depth=3)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
#dinuja el arbol
plt.figure(figsize=(15, 10))
plot_tree(model, feature_names=x.columns, class_names=y.unique(), filled=True, rounded=True)
plt.show()
```

#### <font color='264CC7'> Evaluación del modelo</font>


```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

### <font color='264CC7'> Conclusiones </font>


El árbol de decisión alcanzó una precisión del 80.9% y clasificó con alta efectividad a estudiantes (f1-score: 0.98) y retirados (0.87). Mostró debilidades al identificar doctores (recall: 0.52), lo que indica dificultades para distinguir su patrón bancario. Su estructura simple favorece interpretabilidad, pero compromete el rendimiento en clases con mayor solapamiento.
